```python
# Exercise 1
import pandas as pd

def load_data() -> pd.DataFrame:
    """
    A function that accesses the file on Tesla stock price history on the course website.
    Returns that data as a pd.DataFrame
    """
    temp_df = pd.read_csv("https://lukashager.netlify.app/econ-481/data/TSLA.csv", index_col=0, parse_dates=True)
    temp_df.index = pd.to_datetime(temp_df.index)
    
    return temp_df
```

```python
# Exercise 2
import matplotlib.pyplot as plt

def plot_close(df: pd.DataFrame, start: str = '2010-06-29', end: str = '2024-04-15') -> None:
    """
    A function that takes the output of load_data() as defined above, as well as an optional start and end date (strings formatted as ‘YYYY-MM-DD’).
    Plots the closing price of the stock between those dates as a line graph. Include the date range in the title of the graph.
    """


    # Filter the DataFrame to the specified date range
    mask = (df.index >= start) & (df.index <= end)
    filtered_df = df.loc[mask]

    plt.style.use('classic')
    plt.figure(figsize=(12, 6))  # Define the size of the figure
    plt.plot(filtered_df.index, filtered_df['Close'], label='Close Price', color='blue', linewidth=2)
    plt.title(f'Tesla Stock Close Prices from {start} to {end}')  # Set the title with the date range
    plt.xlabel('Date')  # Label for the X-axis
    plt.ylabel('Close Price (USD)')  # Label for the Y-axis
    plt.grid(True)  # Add a grid for better readability
    plt.legend()  # Add a legend
    plt.show() 

```

```python
plot_close(load_data(), '2019-06-29', '2024-04-15')
```

```python
import statsmodels.api as sm
# Exercise 3
def autoregress(df: pd.DataFrame) -> float:
    """
    This function takes a dataframe as argument.
    Return t-statistics on beta0 hat from regression of delta xt to delta xt - 1
    Using HC1 standard errors in regression
    """
    # Calculate \Delta x_t
    df['delta_close'] = df['Close'].diff()
    
    # Create the lagged variable \Delta x_{t-1}
    df['lag_delta_close'] = df['delta_close'].shift(1)
    
    # Remove the rows with NaN values that result from differencing and shifting
    df.dropna(inplace=True)
    
    # Define the independent variable (lagged differences) and dependent variable (current differences)
    X = df['lag_delta_close']
    y = df['delta_close']

    # Create the OLS model without a constant
    model = sm.OLS(y, X).fit(cov_type='HC1')  # Use heteroscedasticity-corrected standard errors (HC1)
    
    return model.tvalues[0] # Return the t-statistic for \hat{\beta}_0
```

```python
autoregress(load_data())
```

```python
# Exercise 4
def autoregress_logit(df: pd.DataFrame) -> float:
    """
    This function takes a dataframe as argument.
    Return t-statistics on beta0 hat from logistic regression of positive delta xt
    """
    # Calculate \Delta x_t
    df['delta_close'] = df['Close'].diff()

    df['delta_Positive'] = (df['delta_close'] > 0).astype(int)
    
    # Create the lagged variable \Delta x_{t-1}
    df['lag_delta_close'] = df['delta_close'].shift(1)
    
    # Remove the rows with NaN values that result from differencing and shifting
    df.dropna(inplace=True)
    
    # Define the independent variable (lagged differences) and dependent variable (current differences)
    X = df['lag_delta_close']
    y = df['delta_Positive']

    # Create the OLS model without a constant
    model = sm.Logit(y, X).fit(disp=0) 
    
    return model.tvalues['lag_delta_close'] # Return the t-statistic for \hat{\beta}_0
```

```python
autoregress_logit(load_data())
```

```python
# Exercise 5
def plot_delta(df: pd.DataFrame) -> None:
    """
    This function takes a single argument df and plots changes in closing price for the full dataset.
    """
    df['delta_close'] = df['Close'].diff()
    plt.style.use('classic')
    plt.figure(figsize=(12, 6))  # Define the size of the figure
    plt.plot(df.index, df['delta_close'], label=r'$\Delta x_t$', color='blue', linewidth=2)
    plt.grid(color='gray', linestyle=':', linewidth=0.5)
    plt.title(r'$\Delta x_t$: Changes in Tesla Stock Market Price')
    plt.xlabel('Year')
    plt.ylabel('Change in Close Price (USD)')
    plt.legend()
    plt.show() 
    
```

```python
plot_delta(load_data())
```

```python
# idk
def autoregress2(df: pd.DataFrame) -> float:
    """
    Perform a regression on changes in consecutive trading days' closing prices.
    """
    df['lag_close'] = df['Close'].shift(1, freq='B')
    df['delta'] = df['Close'] - df['lag_close']
    df['lag_delta'] = df['delta'].shift(1)
    
    df.dropna(inplace=True)

    # Setup for regression without intercept
    X = df['lag_delta']
    y = df['delta']
    model = sm.OLS(y, X).fit(cov_type='HC1')

    return model.tvalues[0]
```

```python
autoregress2(load_data())
```
